```python
import matplotlib.pyplot as plt
import numpy as np

t = np.arange(0, 2*np.pi, 0.1)
x = 16*np.sin(t)**3
y = 13*np.cos(t)-5*np.cos(2*t)-2*np.cos(3*t)-np.cos(4*t)
#x = 1*(2*np.cos(t)-np.cos(2*t))
#y = 1*(2*np.sin(t)-np.sin(2*t))

plt.title("Leilei")
plt.axis('off')
plt.plot(x, y, color='r')
plt.show()
```
